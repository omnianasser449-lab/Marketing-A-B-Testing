```python
import pandas as pd
```

```python
url=('/content/sql-results.csv')
df= pd.read_csv(url)
print(df)
```

```python
total_conversions = [4695,201]
total_users = [89180, 6718]
```

```python
from statsmodels.stats.proportion import proportions_ztest
z_state, p_value = proportions_ztest(total_conversions, total_users)
print(f"The P-Value is: {p_value:.15f}")

if p_value < 0.05:
    print("The Ads are better.")
else:
    print("We might have just been lucky.")
```
